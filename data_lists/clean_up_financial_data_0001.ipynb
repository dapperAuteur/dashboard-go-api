```python
import pandas as pd
```

```python
tranx_d = pd.read_csv("./tranx_ready - Copy of tranx_ready.csv")
```

```python
vendors_d = pd.read_csv("./mongo_data/vendors_mongo.csv")
```

```python
# create dictionary
# add data[k:v]
vendors_dict = {}
for k,v in vendors_d.iterrows():
    t = v.vendor_name.strip()
    vendors_dict[t] = v._id
```

```python
for k,v in tranx_d.iterrows():
#     print("v.vendor: ",v.vendor)
    
#     print("t_vendor: ",t_vendor)
    
#     print("i_vendor: ",i_vendor)
#     print("vendors_dict[t_vendor]: ",vendors_dict[t_vendor])
    t_vendor = v.vendor
    i_vendor = vendors_dict[t_vendor]
    v.vendor_id = i_vendor
```
